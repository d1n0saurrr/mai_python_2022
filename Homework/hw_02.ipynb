 ЗАДАНИЕ 2: Расчет статистик

```python
dicti = {}
for pilot in pilots:
  dicti[pilot] = len(pilots[pilot]['missions'])

print(f"Пилоты в порядке убывания количества миссий: {dict(sorted(dicti.items(), key=lambda item: item[1], reverse = True))}")
```

```python
sett = set()
for pilot in pilots:
  for mission in pilots[pilot]['missions']:
    sett.add(mission['drone'])
print(f'Полеты совершались на дронах следующих моделей: {", ".join(sett)}')
```

```python
dicti2 = {}
for pilot in pilots:
  for mission in pilots[pilot]['missions']:
    if mission['drone'] in dicti2:
        dicti2[mission['drone']] +=1
    else:
        dicti2[mission['drone']] = 1

for key, val in dicti2.items():
  print(f'Дрон {key} отлетал {val} миссий')
```

ЗАДАНИЕ 3: Создание классов

```python
class Aircraft:
	def __init__(self, weight):
		self._weight = weight
```

```python
class UAV:
  def __init__(self):
    self._has_autopilot = True
    self._missions = []

  @property
  def missions(self):
    return self._missions

  @missions.setter
  def missions(self, mission):
    self._missions.append(mission)

  def count_missions(self):
    return len(self._missions)
```

```python
class MultirotorUAV(Aircraft, UAV):
  def __init__(self, weight, model, brand):
    super().__init__(weight)
    UAV.__init__(self)
    self.__weight = weight
    self.__model = model
    self.__brand = brand

  def get_info(self):
    return f'Информация о дроне {self.get_model()}: масса {self.__weight}, производитель {self.__brand}, количество миссий {self.count_missions()}'

  def get_model(self):
    return self.__model
```

```python
drone_catalog = {
	"DJI Mavic 2 Pro": {"weight":903, "brand":"DJI"},
	"DJI Mavic 2 Zoom": {"weight":900, "brand":"DJI"},
	"DJI Mavic 2 Enterprise Advanced": {"weight":920, "brand":"DJI"},
	"DJI Inspire 2": {"weight":1500, "brand":"DJI"},
	"DJI Mavic 3": {"weight":1000, "brand":"DJI"}
}
```

```python
arr = []
for pilot in pilots:
  for mission in pilots[pilot]['missions']:
    model = mission['drone']
    a = MultirotorUAV(drone_catalog[model]['weight'], model, drone_catalog[model]['brand'])
    for miss in mission['mission']:
      a.missions = miss
    arr.append(a)
```

```python
user_input = input("Введите модель дрона (полностью) в любом регистре\n")
for a in arr:
  if a.get_model().lower() == user_input.lower():
    print(a.get_info())
    break
```
